```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
image_url = 'https://ctrlgeekpod.com/wp-content/uploads/2013/11/attack-on-titan-banner-1.png?w=402&h=306'
```

```python
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)
```

```python
image = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Image from URL")
plt.show()
```

## Task 1

Extract a 100×100 Region from the Center of the Image read from an URL.



```python
height, width = img_rgb.shape[:2]
center_x = width // 2
center_y = height // 2

region = img_rgb[center_y - 50:center_y + 50, center_x - 50:center_x + 50]
```

```python
plt.imshow(region)
plt.axis('off')
plt.title("Cropped Image")
plt.show()
```

## Task - 2
Convert the image to HSV color space, visualize each components in grayscale representation.

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
hue, saturation, value = cv2.split(hsv_image)

plt.figure(figsize=(10, 3))
plt.subplot(131)
plt.imshow(hue, cmap='gray')
plt.title('Hue')
plt.subplot(132)
plt.imshow(saturation, cmap='gray')
plt.title('Saturation')
plt.subplot(133)
plt.imshow(value, cmap='gray')
plt.title('Value')
plt.show()
```
